```python
import cv2
import numpy as np
```

```python
img = cv2.imread("/content/dog.jpg")
```

```python
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
range1 = (30,0,100)
range2 = (255,255,255)
mask = cv2.inRange(hsv,range1,range2)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7))
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
```

```python
mask = cv2.merge([mask,mask,mask])
```

```python
mask_inv = 255 - mask
```

```python
white = np.full_like(img, (255,255,255))
```

```python
img_masked = cv2.bitwise_and(img, mask)
```

```python
white_masked = cv2.bitwise_and(white, mask_inv)
```

```python
result = cv2.add(img_masked, mask_inv)
```

```python
cv2.imwrite("pic_mask.png", mask)
cv2.imwrite("white_background.jpg", result)
```
